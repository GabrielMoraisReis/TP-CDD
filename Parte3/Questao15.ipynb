### Comparação da relação entre as cores dos estudantes nas instituições públicas e privadas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
alumn_column = ["TP_COR_RACA", "TP_CATEGORIA_ADMINISTRATIVA"]
alunos = pd.read_csv(r"C:\Users\arthu\Desktop\TpDados\Aluno.csv", index_col=False, header=0, usecols=alumn_column )
```

```python
alunos["TP_COR_RACA"] = alunos.TP_COR_RACA.replace([1,2,3,4,5], ["Branca", "Preta", "Parda", "Amarela", "Indígena"])
```

```python
alunosParticular = alunos[(alunos.TP_CATEGORIA_ADMINISTRATIVA == 4) | (alunos.TP_CATEGORIA_ADMINISTRATIVA == 5)
                             | (alunos.TP_CATEGORIA_ADMINISTRATIVA == 6)  | (alunos.TP_CATEGORIA_ADMINISTRATIVA == 8)
                              | (alunos.TP_CATEGORIA_ADMINISTRATIVA == 9)]
```

```python
print(alunosParticular['TP_COR_RACA'].value_counts().to_dict())

seaborn.countplot(x='TP_COR_RACA', data=alunosParticular)
plt.title("Particular")
plt.show()
```

```python
alunosPublico = alunos[(alunos.TP_CATEGORIA_ADMINISTRATIVA == 1) | (alunos.TP_CATEGORIA_ADMINISTRATIVA == 2)
                            | (alunos.TP_CATEGORIA_ADMINISTRATIVA == 3)]
```

```python
print(alunosPublico['TP_COR_RACA'].value_counts().to_dict())

seaborn.countplot(x='TP_COR_RACA', data=alunosPublico)
plt.title("Público")
plt.show()
```
